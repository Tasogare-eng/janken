```python
import random

COMPUTER = 0
HAND = ["グー", "チョキ", "パー"]

#　入力の数字化と例外処理
def player_input():
    while True:
        try:
            player = int(input("0.グー、1.チョキ、2.パー（数字を選択して下さい)："))
            if player in [0, 1, 2]:
                break
            else:
                print("数字の0, 1, 2を選択して下さい。")
        except:
            print("数字の0, 1, 2を選択して下さい。")
    return player

def computer_input():
    computer = random.randint(0, 2)
    return computer

# 判定処理を数字化
WIN_TABLE = [[0, 1, -1], [-1, 0, 1], [1, -1, 0]]
def win_lose(player, computer):
    print(f"あなた:{HAND[player]} vs コンピュータ:{HAND[computer]}")
    judgement = WIN_TABLE[player][computer]
    if judgement == 0:
        msg = "引き分け"
    elif judgement == 1:
        msg = "プレーヤーの勝ち"
    elif judgement == -1:
        msg = "コンピュターの勝ち"
    else:
        msg = "入力に誤りがあります"
    print(msg)
    return judgement

print("ゲームを開始します")

# _ で回数確認
# i でカウント
# 0からとなるの＋１
# カウント処理を入れる
# judgementで勝ち負けを受け取りカウントする
player_win = 0
computer_win = 0
for i in range(3):
    print(f"{i+1}回目の勝負です。")
    player = player_input()
    computer = computer_input()
    judgement = win_lose(player, computer)
    if judgement > 0:
        player_win += 1
    elif judgement < 0:
        computer_win += 1

if player_win == computer_win:
    print(f"同点のため最終決戦を行います。")
    while True:
        player = player_input()
        computer = computer_input()
        judgement = win_lose(player, computer)
        if judgement != 0:
            break
        print("引き分けです！もう一回！")
    if judgement > 0:
        print("プレイヤーが優勝しました")
    else:
        print("コンピュータが優勝しました")
elif judgement > 0:
    print("プレイヤーが優勝しました")
else:
    print("コンピュータが優勝しました")

print("ゲームを終了します")
```
